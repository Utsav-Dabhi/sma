```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create a sample graph
G = nx.karate_club_graph()

# Compute the adjacency matrix of the graph
adj_matrix = nx.adjacency_matrix(G).todense()

# Apply K Means Clustering
k = 2
kmeans = KMeans(n_clusters=k, random_state=0).fit(adj_matrix)

# Assign each node to a cluster
clusters = kmeans.labels_

# Plot the graph with nodes colored by cluster
pos = nx.spring_layout(G)
plt.figure(figsize=(8, 6))
nx.draw_networkx_nodes(G, pos, node_size=200, cmap=plt.cm.Set1, node_color=clusters)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)
plt.axis('off')
plt.title('K Means Clustering for Social Network Analysis')
plt.show()
```

```python
import networkx as nx
from networkx.algorithms.community.centrality import girvan_newman
import matplotlib.pyplot as plt

# Generate a test graph
G = nx.karate_club_graph()

# Use Girvan-Newman algorithm to detect communities
comp = girvan_newman(G)

# Convert the iterator to a list of communities
communities = []
for c in next(comp):
    communities.append(list(c))

# Generate the node colors for visualization
colors = ['r', 'b', 'g', 'c', 'm', 'y', 'k']

# Draw the graph with nodes colored by community
pos = nx.spring_layout(G)
i = 0
for com in communities:
    nx.draw_networkx_nodes(G, pos, nodelist=list(com), node_color=colors[i])
    i += 1
nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.show()
```
