```python
import pandas as pd
import numpy as np
import random
from faker import Faker

# Generate fake data
fake = Faker()
data = {'Country': []}

for i in range(10):
    data['Country'].append(fake.country())

# Convert the data dictionary into a pandas dataframe
df = pd.DataFrame(data)
df
```

```python
tweet_count_by_country = df.groupby('Country')['Country'].count().reset_index(name='Count')
tweet_count_by_country
```

```python
tweets_count_by_country = df['Country'].value_counts().reset_index(name='Count').rename(columns={'index': 'Country'})
tweets_count_by_country
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.barplot(x='Country', y='Count', data=tweet_count_by_country)
plt.title('Number of Tweets by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```
