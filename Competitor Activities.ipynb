```python
import tweepy
import pandas as pd

# Twitter API credentials
consumer_key = "your_consumer_key"
consumer_secret = "your_consumer_secret"
access_token = "your_access_token"
access_token_secret = "your_access_token_secret"

# Authenticate with Twitter API
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)

# List of competitors Twitter usernames
competitors = ['competitor1', 'competitor2', 'competitor3']

# Gather data from competitor Twitter accounts
tweets = []
for competitor in competitors:
    competitor_tweets = api.user_timeline(screen_name=competitor, count=200, tweet_mode="extended")
    for tweet in competitor_tweets:
        tweets.append([competitor, tweet.id_str, tweet.created_at, tweet.full_text, tweet.favorite_count, tweet.retweet_count])

# Create pandas DataFrame from the gathered data
df = pd.DataFrame(tweets, columns=['Competitor', 'Tweet ID', 'Created At', 'Text', 'Favorites', 'Retweets'])

# Calculate total engagement (favorites + retweets) for each tweet
df['Engagement'] = df['Favorites'] + df['Retweets']

# Calculate average engagement for each competitor
avg_engagement = df.groupby('Competitor')['Engagement'].mean()

# Calculate total engagement for each competitor
total_engagement = df.groupby('Competitor')['Engagement'].sum()

# Calculate total number of tweets for each competitor
total_tweets = df.groupby('Competitor')['Text'].count()

# Print the results
print("Average engagement:\n", avg_engagement)
print("Total engagement:\n", total_engagement)
print("Total tweets:\n", total_tweets)
```

```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Load the dataset
df = pd.read_csv('competitor_tweets.csv')

# Instantiate the SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Add a column for sentiment scores
df['sentiment_score'] = df['text'].apply(lambda x: sia.polarity_scores(x)['compound'])

# Group the tweets by company and calculate the mean sentiment score
grouped = df.groupby('company')['sentiment_score'].mean()

# Print the results
print(grouped)
```
